```python
#This is still a work in progress.



#calling in libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import csv
import seaborn as sns

load_1 = r"C:\Users\mjeas\Desktop\git_repos\Homework\Week_08_Homework\Military_Data2\Active_Duty_Enlisted_2.csv"
#active duty enlisted 
active_enlisted_df = pd.read_csv(load_1)

#active duty officer
load_2 = r"C:\Users\mjeas\Desktop\git_repos\Homework\Week_08_Homework\Military_Data2\Active_Duty_Officers_2.csv"
active_officer_df = pd.read_csv(load_2)

#reserve enlisted
load_3 = r"C:\Users\mjeas\Desktop\git_repos\Homework\Week_08_Homework\Military_Data2\Reserve_Enlisted_2.csv"
reserve_enlisted_df = pd.read_csv(load_3)

#reserve officer
load_4= r"C:\Users\mjeas\Desktop\git_repos\Homework\Week_08_Homework\Military_Data2\Reserve_Officers_2.csv"
reserve_officer_df = pd.read_csv(load_4)
```

```python
#Army Enlisted Active
army_active_en = active_enlisted_df[active_enlisted_df.Branch == "Army"] 

#Army Enlisted National Guard
army_en_ng = reserve_enlisted_df[reserve_enlisted_df.Branch == "Army National Guard"]

#Army Enlisted Reserve
army_reserve_en = reserve_enlisted_df[reserve_enlisted_df.Branch == "Army Reserve"]

#Army Officer Active
army_active_of = active_officer_df[active_officer_df.Branch == "Army"]

#Army Officer National Guard
army_of_ng = reserve_officer_df[reserve_officer_df.Branch == "Army National Guard"]

#Army Officer Reserve
army_of_reserve = reserve_officer_df[reserve_officer_df.Branch == "Army Reserve"]

#Total Army
total_army_df = army_active_en.append([army_en_ng, army_reserve_en,army_active_of,army_of_ng, army_of_reserve])
```

```python
#What is the racial breakdown of all Army Components?
total_army_df["Race Totals"]= total_army_df.sum(axis =1,)
race_sum_army = total_army_df.groupby("Race").sum()
#total_soldiers = race_sum_army.shape[0]      
#race_percentage_soldiers = (race_sum_army["Race Totals"] /total_soldiers["Race Totals"])* 100


fig, ax = plt.subplots()
ax.bar(race_sum_army.index, race_sum_army["Race Totals"])
ax.set_ylabel("Number of Soldiers")
plt.show()
```

```python
#Which Component is the most racially diverse?

#Active
army_active = army_active_en.append([army_active_of])
army_active["Race Totals"]= army_active.sum(axis =1,)
race_sum_army_active = army_active.groupby("Race").sum().plot(kind='pie', y='Race Totals',autopct='%1.1f%%')

#National Guard
army_ng = army_en_ng.append([army_of_ng])
army_ng["Race Totals"]= army_ng.sum(axis =1,)
race_sum_army_ng = army_ng.groupby("Race").sum().plot(kind='pie', y='Race Totals', autopct='%1.1f%%')

#Reserve
army_res = army_reserve_en.append([army_of_reserve])
army_res["Race Totals"]= army_res.sum(axis =1,)
race_sum_army_res = army_res.groupby("Race").sum().plot(kind='pie', y='Race Totals', autopct='%1.1f%%')


```
